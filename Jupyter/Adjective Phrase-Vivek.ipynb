```python
import pandas as pd
import spacy
from collections import Counter
nlp = spacy.load("en_core_web_sm")
df = pd.read_csv('../Data/review.csv')
```

```python
reviews = df[df['business_id'] == 'j7HO1YeMQGYo3KibMXZ5vg'][['text']]
reviews = reviews.reset_index().drop(columns='index')
reviews.head()
```

## Capture Type 1 of Adjective Phrases:

```python
adv_adj_pairs_type1 = []

for j in range(len(reviews)):
    doc = nlp(reviews.iloc[j]["text"])
    pairs = []
    flag = -1
    for i,token in enumerate(doc):
        if token.pos_ == 'ADV':
            pairs = []
            pairs.append(j)
            pairs.append(i)
            pairs.append(token)
            flag = i
#             print(token,": ",token.pos_,": ",i)
#             print("Found Adverb:",pairs)
        if token.pos_ == 'ADJ':
            if(i == flag+1 and flag!=-1):
                pairs.append(token)
#                 print(token,": ",token.pos_,": ",i)
#                 print("Found Adjective:",pairs)
                adv_adj_pairs_type1.append(pairs)
                pairs = []
#                 print(adv_adj_pairs)
                flag = -1
print(adv_adj_pairs_type1)
       
# out = []

# for pairs in adv_adj_pairs:
#     item_1 = pairs[0]
#     item_2 = pairs[1]
#     out.append((str(item_1), str(item_2)))

# c = Counter(out)
# c.most_common(20)
```

## Capture Type 2 of Adjective Phrases:

```python
adv_adj_pairs_type2 = []

for j in range(len(reviews)):
    doc = nlp(reviews.iloc[j]["text"])
    pairs = []
    flag = -1
    for i,token in enumerate(doc):
        if token.pos_ == 'ADJ':
            pairs = []
            pairs.append(j)
            pairs.append(i)
            pairs.append(token)
            flag = i
#             print(token,": ",token.pos_,": ",i)
#             print("Found Adjective:",pairs)
        if token.pos_ == 'ADP':
            if(i == flag+1 and flag!=-1):
                pp = ' '.join([tok.orth_ for tok in token.subtree])
                pairs.append(pp)
#                 print(token,": ",token.pos_,": ",i)
#                 print("Found Prepostional Phrase:",pairs)
                adv_adj_pairs_type2.append(pairs)
                pairs = []
#                 print(adv_adj_pairs_type2)
                flag = -1
print(adv_adj_pairs_type2)
```

## Capture Type 3 of Adjective Phrases:

```python
adv_adj_pairs_type3 = []

for j in range(len(reviews)):
    doc = nlp(reviews.iloc[j]["text"])
    pairs = []
    flag1 = -1
    flag2 = -1
    for i,token in enumerate(doc):
        if token.pos_ == 'ADV':
            pairs = []
            pairs.append(j)
            pairs.append(i)
            pairs.append(token)
            flag1 = i
#             print(token,": ",token.pos_,": ",i)
#             print("Found Adverb:",pairs)
        if token.pos_ == 'ADJ':
            if(i == flag1+1 and flag1!=-1):
                pairs.append(token)
#                 print(token,": ",token.pos_,": ",i)
#                 print("Found Adjective:",pairs)
                flag2 = i
        if token.pos_ == 'ADP':
            if(i == flag2+1 and flag2!=-1 and flag1!=-1):
                pp = ' '.join([tok.orth_ for tok in token.subtree])
                pairs.append(pp)
#                 print(token,": ",token.pos_,": ",i)
#                 print("Found Prepostional Phrase:",pairs)
                adv_adj_pairs_type3.append(pairs)
                pairs = []
#                 print(adv_adj_pairs_type3)
                flag1 =-1
                flag2 = -1
print(adv_adj_pairs_type3)
```

## Removing Duplicates:
1. Type 1 and Type 3 will have overlaps as they are of the form ADV,ADJ and ADV,ADJ,PP respectively
1. Type 2 and Type 3 will have overlaps as they are of the from ADJ,PP and ADV,ADJ,PP respectively

### So we identify the duplicates and remove them from type 1 and type 2 lists as type 3 will have a more complete adjective phrase 

```python
duplicatesInType1 = []
for i in range(len(adv_adj_pairs_type1)):
    for j in range(len(adv_adj_pairs_type3)):
        if(adv_adj_pairs_type1[i][0]==adv_adj_pairs_type3[j][0] and adv_adj_pairs_type1[i][1]==adv_adj_pairs_type3[j][1]):
            print(adv_adj_pairs_type1[i])
            print(adv_adj_pairs_type3[j])
            duplicatesInType1.append(i)
print(duplicatesInType1)
            
```

```python
duplicatesInType2 = []
for i in range(len(adv_adj_pairs_type2)):
    for j in range(len(adv_adj_pairs_type3)):
        if(adv_adj_pairs_type2[i][0]==adv_adj_pairs_type3[j][0] and adv_adj_pairs_type2[i][1]==(adv_adj_pairs_type3[j][1]+1)):
            print(adv_adj_pairs_type2[i])
            print(adv_adj_pairs_type3[j])
            duplicatesInType2.append(i)
print(duplicatesInType2)
            
```

```python
print(duplicatesInType2)
for index in sorted(duplicatesInType2, reverse=True):
    del adv_adj_pairs_type2[index]
```

```python
print(duplicatesInType1)
for index in sorted(duplicatesInType1, reverse=True):
    del adv_adj_pairs_type1[index]
```

## Concatenating all types of Adjective Phrases

```python
total = adv_adj_pairs_type1 + adv_adj_pairs_type2 + adv_adj_pairs_type3
total
```

```python
list_of_adj_phrases = []
for i in total:
    phrase = i[2:]
    str_type = []
    for j in phrase:
        if isinstance(j, str):
            str_type.append(j)
        else:
            str_type.append(j.text)
#     print(str_type)
    adj_phrase = ' '.join(str_type)
#     print(adj_phrase)
    list_of_adj_phrases.append(adj_phrase)
    
list_of_adj_phrases
```

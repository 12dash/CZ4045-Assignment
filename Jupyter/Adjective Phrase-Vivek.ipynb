## Capture Type 2 of Adjective Phrases:

```python
adv_adj_pairs_type2 = []

for j in range(len(reviews)):
    doc = nlp(reviews.iloc[j]["text"])
    pairs = []
    flag = -1
    for i,token in enumerate(doc):
        if token.pos_ == 'ADJ':
            pairs = []
            pairs.append(j)
            pairs.append(i)
            pairs.append(token)
            flag = i
#             print(token,": ",token.pos_,": ",i)
#             print("Found Adjective:",pairs)
        if token.pos_ == 'ADP':
            if(i == flag+1 and flag!=-1):
                pp = ' '.join([tok.orth_ for tok in token.subtree])
                pairs.append(pp)
#                 print(token,": ",token.pos_,": ",i)
#                 print("Found Prepostional Phrase:",pairs)
                adv_adj_pairs_type2.append(pairs)
                pairs = []
#                 print(adv_adj_pairs_type2)
                flag = -1
print(adv_adj_pairs_type2)
```

## Creating Function to extract adjective phrases from reviews:

```python
def getPhrasesType1(reviews):
    adv_adj_pairs_type1 = []

    for j in range(len(reviews)):
        doc = nlp(reviews.iloc[j]["text"])
        pairs = []
        flag = -1
        for i,token in enumerate(doc):
            if token.pos_ == 'ADV':
                pairs = []
                pairs.append(j)
                pairs.append(i)
                pairs.append(token)
                flag = i
    #             print(token,": ",token.pos_,": ",i)
    #             print("Found Adverb:",pairs)
            if token.pos_ == 'ADJ':
                if(i == flag+1 and flag!=-1):
                    pairs.append(token)
    #                 print(token,": ",token.pos_,": ",i)
    #                 print("Found Adjective:",pairs)
                    adv_adj_pairs_type1.append(pairs)
                    pairs = []
    #                 print(adv_adj_pairs)
                    flag = -1
#     print(adv_adj_pairs_type1)
    return adv_adj_pairs_type1

def getPhrasesType2(reviews):
    adv_adj_pairs_type2 = []

    for j in range(len(reviews)):
        doc = nlp(reviews.iloc[j]["text"])
        pairs = []
        flag = -1
        for i,token in enumerate(doc):
            if token.pos_ == 'ADJ':
                pairs = []
                pairs.append(j)
                pairs.append(i)
                pairs.append(token)
                flag = i
    #             print(token,": ",token.pos_,": ",i)
    #             print("Found Adjective:",pairs)
            if token.pos_ == 'ADP':
                if(i == flag+1 and flag!=-1):
                    pp = ' '.join([tok.orth_ for tok in token.subtree])
                    pairs.append(pp)
    #                 print(token,": ",token.pos_,": ",i)
    #                 print("Found Prepostional Phrase:",pairs)
                    adv_adj_pairs_type2.append(pairs)
                    pairs = []
    #                 print(adv_adj_pairs_type2)
                    flag = -1
#     print(adv_adj_pairs_type2)
    return adv_adj_pairs_type2

def getPhrasesType3(reviews):
    adv_adj_pairs_type3 = []

    for j in range(len(reviews)):
        doc = nlp(reviews.iloc[j]["text"])
        pairs = []
        flag1 = -1
        flag2 = -1
        for i,token in enumerate(doc):
            if token.pos_ == 'ADV':
                pairs = []
                pairs.append(j)
                pairs.append(i)
                pairs.append(token)
                flag1 = i
    #             print(token,": ",token.pos_,": ",i)
    #             print("Found Adverb:",pairs)
            if token.pos_ == 'ADJ':
                if(i == flag1+1 and flag1!=-1):
                    pairs.append(token)
    #                 print(token,": ",token.pos_,": ",i)
    #                 print("Found Adjective:",pairs)
                    flag2 = i
            if token.pos_ == 'ADP':
                if(i == flag2+1 and flag2!=-1 and flag1!=-1):
                    pp = ' '.join([tok.orth_ for tok in token.subtree])
                    pairs.append(pp)
    #                 print(token,": ",token.pos_,": ",i)
    #                 print("Found Prepostional Phrase:",pairs)
                    adv_adj_pairs_type3.append(pairs)
                    pairs = []
    #                 print(adv_adj_pairs_type3)
                    flag1 =-1
                    flag2 = -1
#     print(adv_adj_pairs_type3)
    return adv_adj_pairs_type3

def removeDuplicatesType1(adv_adj_pairs_type1,adv_adj_pairs_type3):
    duplicatesInType1 = []
    for i in range(len(adv_adj_pairs_type1)):
        for j in range(len(adv_adj_pairs_type3)):
            if(adv_adj_pairs_type1[i][0]==adv_adj_pairs_type3[j][0] and adv_adj_pairs_type1[i][1]==adv_adj_pairs_type3[j][1]):
#                 print(adv_adj_pairs_type1[i])
#                 print(adv_adj_pairs_type3[j])
                duplicatesInType1.append(i)
#     print(duplicatesInType1)

    for index in sorted(duplicatesInType1, reverse=True):
        del adv_adj_pairs_type1[index]
#     print("Length of Type 1 After Deletion:",len(adv_adj_pairs_type1))
#     print(adv_adj_pairs_type1)
    return adv_adj_pairs_type1

def removeDuplicatesType2(adv_adj_pairs_type2,adv_adj_pairs_type3):
    duplicatesInType2 = []
    for i in range(len(adv_adj_pairs_type2)):
        for j in range(len(adv_adj_pairs_type3)):
            if(adv_adj_pairs_type2[i][0]==adv_adj_pairs_type3[j][0] and adv_adj_pairs_type2[i][1]==(adv_adj_pairs_type3[j][1]+1)):
#                 print(adv_adj_pairs_type2[i])
#                 print(adv_adj_pairs_type3[j])
                duplicatesInType2.append(i)
#     print(duplicatesInType2)
    for index in sorted(duplicatesInType2, reverse=True):
        del adv_adj_pairs_type2[index]
#     print("Length of Type 2 After Deletion:",len(adv_adj_pairs_type2))
#     print(adv_adj_pairs_type2)
    return adv_adj_pairs_type2

def returnAdjectivePhrases(adv_adj_pairs_type1,adv_adj_pairs_type2,adv_adj_pairs_type3):
    total = adv_adj_pairs_type1 + adv_adj_pairs_type2 + adv_adj_pairs_type3
    list_of_adj_phrases = []
    for i in total:
        phrase = i[2:]
        str_type = []
        for j in phrase:
            if isinstance(j, str):
                str_type.append(j)
            else:
                str_type.append(j.text)
    #     print(str_type)
        adj_phrase = ' '.join(str_type)
    #     print(adj_phrase)
        list_of_adj_phrases.append(adj_phrase.lower())
    return list_of_adj_phrases

def generateAdjectivePhrases(reviews):
    adv_adj_pairs_type1 = getPhrasesType1(reviews)
    adv_adj_pairs_type2 = getPhrasesType2(reviews)
    adv_adj_pairs_type3 = getPhrasesType3(reviews)
    adv_adj_pairs_type1 = removeDuplicatesType1(adv_adj_pairs_type1,adv_adj_pairs_type3)
    adv_adj_pairs_type2 = removeDuplicatesType2(adv_adj_pairs_type2,adv_adj_pairs_type3)
    return returnAdjectivePhrases(adv_adj_pairs_type1,adv_adj_pairs_type2,adv_adj_pairs_type3)
```

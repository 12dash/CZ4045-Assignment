```python
random_reviews = []
random_reviews.append(review_text[69])
random_reviews.append(review_text[420])
random_reviews.append(review_text[0])
random_reviews.append(review_text[1056])
random_reviews.append(review_text[11604])
```

```python
random_reviews
```

## NLTK

```python
 nltk.download('averaged_perceptron_tagger')
```

```python
pos_tag_list = []

for review in random_reviews:
  tokens = word_tokenize(review)
  pos_tag_list.append(nltk.pos_tag(tokens))


print("Pos Tag List \n", (pos_tag_list))

```

## Spacy

<!-- https://spacy.io/usage/linguistic-features -->
<!-- https://machinelearningknowledge.ai/tutorial-on-spacy-part-of-speech-pos-tagging/ -->
<!-- For example, en_core_web_sm is a small English pipeline trained on written web text (blogs, news, comments), that includes vocabulary, syntax and entities. -->

```python
import spacy
import en_core_web_sm
temp =set()
nlp = en_core_web_sm.load()
pos_tag_list_spacy = []
for review in random_reviews:
  doc = nlp(review)
  tmp = []
  for token in doc:
    temp = (token.text, token.tag_)
  #  temp = (token.text, token.pos_)
    tmp.append(temp)
  
  pos_tag_list_spacy.append(tmp)
```

```python
import nltk
import pandas as pd
import numpy as np

from nltk.tokenize import word_tokenize 
```

```python
# Read from csv
rev_df = pd.read_csv('../Data/review.csv')
rev_df.head()
```

```python
len(rev_df['business_id'].unique())
unique_ids = list(rev_df['business_id'].unique())
```

```python
df_new = rev_df[rev_df['stars']  == 1.0]
```

```python
df_new
reviews_1star = []

for index,row in df_new.iterrows():
  if (row['business_id'] in unique_ids) and (len(reviews_1star) <50):
    reviews_1star.append(row['text'])
    i = unique_ids.index(row['business_id'])
    unique_ids[i] = -1

```

```python
len(reviews_1star)
```

```python
import spacy
import en_core_web_sm
nlp = en_core_web_sm.load()
noun_adj_pairs = []
for review in reviews_1star:
  doc = nlp(review)
  noun_adj_pairs_temp = []
  for i,token in enumerate(doc):
      if token.pos_ not in ('NOUN','PROPN'):
          continue
      for j in range(i+1,len(doc)):
          if doc[j].pos_ == 'ADJ':
              noun_adj_pairs.append((token,doc[j]))
              break
  noun_adj_pairs.append(noun_adj_pairs_temp)
noun_adj_pairs
```
